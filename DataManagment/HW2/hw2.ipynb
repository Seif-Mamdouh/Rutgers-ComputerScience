```python
# Question 1
# Write a program that reads words from a filename, which is given as a string argument. It should return the words from the file in a list, sorted in reverse alphabetical order(case insensitive) .

# For instance, if the file has

# bell tea  Zebra
# apple
# yellow
# Then the output should be

['Zebra', 'yellow', 'tea', 'bell', 'apple']


def reverse_sorted_words(filename):
    # Read words from the file and split them into a list
    with open(filename, 'r') as file:
        words = file.read().split()

        sorted_words = sorted(words, key=str.lower, reverse=True)

    return sorted_words


# Use this code block to test your program, please make your own tests too.
answer_key = ['whispering', 'visitor', 'utopian', 'Thrill', 'tangible', 'ray', 'point', 'pan', 'launch',
              'hurry', 'houses', 'frog', 'five', 'fang', 'continue', 'buzz', 'Ask', 'airport', 'advise', 'activity']
rev_words = reverse_sorted_words("words.txt")
assert len(answer_key) == len(rev_words)
for ans, res in zip(answer_key, rev_words):
    # check if each word matches the answer key
    assert ans == res
```

```python
# Write a program that reads words from a file (filename given as a string parameter) and prints the occurance of each word(case insensitive). And print the words in alphabetical order.

# For example, if the file contains text

# Love is free
# free is love
# then the function should print

# free:2
# is:2
# love:2


def count_word(filename):
    word_count = {}  # Initialize an empty dictionary to store word counts
    # Open the file for reading
    with open(filename, 'r') as file:
        for line in file:
            words = line.strip().split()
            for word in words:
                word = word.strip('.,?!-()[]{}:;\'"').lower()
                if word not in word_count:
                    word_count[word] = 1
                else:
                    # If the word is already in the dictionary, increment its count
                    word_count[word] += 1

    # Sort the words lexigraphacally
    sorted_words = sorted(word_count.keys())

    # Print the words and their counts
    for word in sorted_words:
        print(f"{word}:{word_count[word]}")​
# # Example usage:
# count_word('words2.txt')
# activity:1
# advise:2
# airport:1
# ask:1
# buzz:5
# continue:2
# fang:1
# five:2
# frog:2
# houses:3
# hurry:3
# launch:1
# pan:1
# point:1
# ray:2
# tangible:1
# thrill:1
# utopian:1
# visitor:3
# whispering:1
```

```python
# Write a program that reads words passing from a string parameter and create a dictionary mapping last letters to the number of words ending with that letter.

# For example, the string is 'apple banana orange grape cherry', the program will out put a dictionary fruitName that

# print(fruitName)
# will give us output

# {'e': 3, 'a': 1, 'y': 1}

def count_last_letter(words):
    word_list = words.split()

    last_letter_count = {}

    # Get the last letter
    for word in word_list:
        last_letter = word[-1].lower()

        if last_letter in last_letter_count:
            # If the letter is already in the dictionary, increment its count
            last_letter_count[last_letter] += 1
        else:
            last_letter_count[last_letter] = 1

    return last_letter_count


fruitName = count_last_letter('apple banana orange grape cherry')
print(fruitName)
```

```python
# Question 5¶
# We can create a numpy array from separate lists by specifying a format for each field:

# group = np.zeros(4, dtype={'names': ('name', 'age', 'major', 'gpa'),
#                             'formats': ('U50', 'i4', 'U4', 'f8')})

# group['name'] = ['Alice', 'Bob', 'Carol', 'Dennis']
# group['age'] = [21, 25, 18, 29]
# group['major'] = ['CS', 'Math', 'Chem', 'Phys']
# group['gpa'] = [3.8, 3.2, 4.0, 3.5]
# where U50 is a string of max length 50, i4 is a 4-byte integer, U4 is a string of max length 4, and f8 is an 8-byte floating-point number. Then group has value

# array([('Alice', 21, 'CS', 3.8), ('Bob', 25, 'Math', 3.2),
#        ('Carol', 18, 'Chem', 4.), ('Dennis', 29, 'Phys', 3.5)],
#       dtype=[('name', '<U50'), ('age', '<i4'), ('major', '<U4'), ('gpa', '<f8')])
# Then, for example, group['name'] will give an array of just the student names.

# Write a program that reads the numpy array group we just created that has this format:

# name, age, major, gpa

# Convert this to a numpy array as shown above. Then use that array to compute and print each of the following:

# The average GPA of all students
# The maximum GPA of students majoring in CS
# The number of students with a GPA over 3.5
# The average GPA of students who are at least 25 years old
# The major that has the highest average GPA among students at most 22 years old
# For example, for the four students listed, your program should print:

# 3.625
# 3.8
# 2
# 3.35
# Chem


import numpy as np

# Create a numpy array 'group' with the given format
group = np.array([('Alice', 21, 'CS', 3.8), ('Bob', 25, 'Math', 3.2),
                  ('Carol', 18, 'Chem', 4.0), ('Dennis', 29, 'Phys', 3.5)],
                  dtype=[('name', '<U50'), ('age', '<i4'), ('major', '<U4'), ('gpa', '<f8')])

everyone_avg_GPA = np.mean(group['gpa'])
print(everyone_avg_GPA)

computer_science_student = group[group['major'] == 'CS']
max_gpa_cs = np.max(computer_science_student['gpa'])
print(max_gpa_cs)

smart_students = len(group[group['gpa'] > 3.5])
print(smart_students)

age_25_avg = np.mean(group[group['age'] >= 25]['gpa'])
print(age_25_avg)

year_22_or_younger = group[group['age'] <= 22]
one_of_a_kind_majors = np.unique(year_22_or_younger['major'])
max_avg_gpa = -1
major_with_max_avg_gpa = ''

for major in one_of_a_kind_majors:
    major_students = year_22_or_younger[year_22_or_younger['major'] == major]
    avg_gpa = np.mean(major_students['gpa'])
    if avg_gpa > max_avg_gpa:
        max_avg_gpa = avg_gpa
        major_with_max_avg_gpa = major

print(major_with_max_avg_gpa)
```
